### IMPORTING PACKAGES AND LIBRARIES

### LOADING THE DATASET

```python
df1=pd.read_csv(r'C:\Users\ADMIN\Downloads\Got eps.csv')
df1.head()
```

### VIEWING DATASET COLUMNS

```python
df1.columns
```

### DROPPING UNNECESSARY COLUMNS
 **The columns 'Summary','Star_1', 'Star_2', 'Star_3','Budget_estimate' are not necessary for visualization and hence removed.**

```python
df1.drop(['Summary','Star_1', 'Star_2', 'Star_3','Budget_estimate'],axis=1)
```

```python
df1.isnull().sum()
```

### VISUALIZATION BASED ON DIRECTORS
**The columns are grouped such that directors and the corresponding reviews are separated out. Those values are plotted using a bargraph.**

```python
season_wise_rating= df1.groupby('Director')[['Users_reviews','Critics_reviews']].sum().reset_index()
season_wise_rating
```

```python

plt.figure(figsize=(10,7))
sns.barplot(season_wise_rating['Director'],season_wise_rating['Users_reviews'])
plt.title("Director Vs Users Reviews",fontsize = 15)
plt.xlabel("Director",fontsize = 10)
plt.ylabel("Users_reviews",fontsize = 10)
plt.xticks(rotation=90)
plt.plot()

plt.figure(figsize=(10,7))
sns.barplot(season_wise_rating['Director'],season_wise_rating['Critics_reviews'])
plt.title("Director Vs Critics Reviews",fontsize = 15)
plt.xlabel("Director",fontsize = 10)
plt.ylabel("Critics_reviews",fontsize = 10)
plt.xticks(rotation=90)
plt.plot()

```

**From the graph above we can see that David Nutter has the highest critics review followed by miguel Sapochnik.Interestingly these two have very high user review as well,with miguel taking lead in the latter.**

### VISUALIZATION FOR DURATION OF EPISODES

```python
plt.figure(figsize=(10,7))
plt1 = sns.distplot(df1['Duration'])
plt.title("Distribution of Duration for various episodes",fontsize=15)
plt.show()
```

**We find that most of the episodes have a average duration of 50 mins-60 mins.Few episodes can extend upto 80 mins as we see in the graph.**

### SEASON NAMES COLUMN


```python
season=[]
for i in df1.Season:
    if(i==1):
        season.append('Season 1')
    elif(i==2):
        season.append('Season 2')
    elif(i==3):
        season.append('Season 3')
    elif(i==4):
        season.append('Season 4')
    elif(i==5):
        season.append('Season 5')
    elif(i==6):
        season.append('Season 6')
    elif(i==7):
        season.append('Season 7')
    elif(i==8):
        season.append('Season 8')
   
        
        
        
df1['Season_name']=season
df1['Season_name']
```

### AVERAGE RATING VS SEASON

```python
season_rating = round(df1.groupby('Season').Rating.mean().to_frame(),2).reset_index()
season_rating.columns = ['Season', 'Rating']
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=season_rating.Season, y=season_rating.Rating,palette="husl")
plt.ylim(5.5,9.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Average Rating Per Season", fontsize=16)
plt.ylabel('Rating', fontsize=14)
plt.xlabel('Season', fontsize=14)
```

**From the graph we can see that season 4 has the highest rating whereas season 8 has lowest rating.This is very interesting considering that Joffrey Baratheon, the greatest villian of all time died during season 4 and Daenerys Targaryen,one of the main protagonist and one of fan favourites died during season 8**

### TOP 10 EPISODES OF GoT

```python
top_10_eps = df1[['Season', 'Episode','Title', 'Rating', 'Votes']].sort_values(['Rating','Votes'], ascending=False)[:10]
top_10_eps
```

**As an ardent fan of Game of Thrones I can definitely say Battle of Bastards is indeed hands-down best episode of the series.**

```python
plt.figure(figsize=(10,5))
sns.barplot(y=top_10_eps.Title, x=top_10_eps.Rating, orient='h', palette="GnBu_d")
plt.xlim(9,10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Top Ten episodes by Ratings", fontsize=15)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Episode Title', fontsize=14)
```

### BOTTOM 10 EPISODES OF GoT

```python
bottom_10_eps = df1[['Season', 'Episode','Title', 'Rating', 'Votes']].sort_values(['Rating','Votes'], ascending=True)[:10]
bottom_10_eps
```

**The least favourite is The Iron Throne from season 8. It is interesting to note that most of the epsisodes from bottom list belongs to season 8.**

```python
plt.figure(figsize=(10,5))
sns.barplot(y=bottom_10_eps.Title, x=bottom_10_eps.Rating, orient='h', palette='Reds_d')
plt.xlim(3,9)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Bottom Ten episodes by Ratings", fontsize=15)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Episode Title', fontsize=14)
```
